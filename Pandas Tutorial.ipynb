# Filtering - Using Conditionals to Filter Rows and Columns

```python
df12
```

```python
df.head()
```

```python
## To identify that whoesever Last Name matches with the last name value. Output is True or False which matches with our data

df12['Last Name']=='Gowda'
```

```python
filt= (df12['Last Name']=='Gowda')
```

```python
df12[filt]
```

```python
df12.loc[filt,'Email']
```

```python
# Using the And operations in the Filter

filt= (df12['Last Name']=='Gowda') & (df12['First Name']=='Vishwas')
```

Point to be noted is if we filter out on some condition and assign it to a variable , we can just see only the Boolean values of it like TRUE or FALSE . In order to see the original values of the result use the data frame along with the filter condition. 

```python
df12[filt]
```

```python
df12.loc[filt,'Email']
```

```python
# ~ filt to negate 

df12.loc[~filt,'Email']
```
